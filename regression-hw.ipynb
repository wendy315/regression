```python
#自變數變換
value3_tr=pd.DataFrame()
value3_tr['log']=np.log(profit_data['value3']) 
value3_tr['log2']=np.log2(profit_data['value3'])
value3_tr['sqrt']=np.sqrt(profit_data['value3']) #平方開根號
value3_tr['1/sqrt']=1/(np.sqrt(profit_data['value3'])) #倒數
value3_tr['reciprocal']=np.reciprocal(profit_data['value3'])
value3_tr['square']=np.power(profit_data['value3'],2) #平方

value3_tr.hist(bins=20,layout=(3,2),edgecolor='k',figsize=(10,8))
plt.suptitle('value3')
plt.show()
```

```python
value3=value3_tr['sqrt']
```

```python
#共線性檢定 vif>看有無10
from statsmodels.stats.outliers_influence import variance_inflation_factor

#加入自變數RD_Spend,Marketing_Spend和常數1
X=sm.add_constant(profit_data.loc[:,['RD_Spend','Marketing_Spend','Administration','value1','value2','value3']]) #看要放甚麼

vif=pd.DataFrame()
vif['features']=X.columns
vif['VIF_Factor']=[variance_inflation_factor(X.values,i) for i in range(X.shape[1])]

vif
```

```python
#計算應變數與自變數的coefficient_of_association
#選擇較高的coe自變數
coe=profit_data.drop('Profit',axis=1).corrwith(profit_data.Profit)
coe
```

```python
#透過視覺化方式
sns.pairplot(profit_data.loc[:,['RD_Spend','Administration','Marketing_Spend','value1','value2','value3','Profit']])
plt.show()
```

```python
#進行回歸模型
from sklearn.model_selection import train_test_split

train,test=train_test_split(profit_data,test_size=0.2,random_state=1) 
model=sm.formula.ols('Profit ~ RD_Spend+Marketing_Spend+value3',data=train).fit()
print(model.params)
```

```python
#Profit=49283.434312+0.749857RD_Spend+0.027830Marketing_Spend+ 0.019457value3
```

```python
model.summary()
#OLS Regression Results最小二乘估計 反正就是線性回歸 rs0.96 x可以解釋y
#p<0.05 RD_Spend比Marketing_Spend重要 永遠選擇最好的
```

```python
#預測值
pred=model.predict(test)
pred
```

```python
#實際值
test.Profit
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error

print("Root mean squared error: %.2f" % np.sqrt(mean_squared_error(test.Profit, pred)))
print("Mean absolute error: %.2f" % mean_absolute_error(test.Profit, pred))
print("Mean absolute percentage error: %.2f" % mean_absolute_percentage_error(test.Profit, pred))
```

# 增加模型複雜度

```python
#相乘
profit_data['RD_Marketing_Spend']=profit_data['RD_Spend']*profit_data['Marketing_Spend']

#平方
profit_data['RD_Spend_2']=profit_data['RD_Spend']**(2)*profit_data['Marketing_Spend']

#次方
profit_data['RD_Spend_3']=profit_data['RD_Spend']**(3)

profit_data.head()

#多出RD_Marketing_Spend	RD_Spend_2	RD_Spend_3
```

```python
train,test=train_test_split(profit_data,test_size=0.2,random_state=1)
model=sm.formula.ols('Profit ~ RD_Spend+Marketing_Spend+RD_Marketing_Spend+RD_Spend_2',data=train).fit() 
pred=model.predict(test)
print(model.params)
```

```python
print("Root mean squared error: %.2f" % np.sqrt(mean_squared_error(test.Profit, pred)))
print("Mean absolute error: %.2f" % mean_absolute_error(test.Profit, pred))
print("Mean absolute percentage error: %.2f" % mean_absolute_percentage_error(test.Profit, pred))
#更擬和一點 衍生變數 目的複雜化 刪掉勝好變數 用好變數再增添 複雜化
```
